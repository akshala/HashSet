```python
import pandas as pd
```

```python
def is_camel_case(s):
    capitalCount = len([char for char in s if char.isupper()])

    if capitalCount == 1 and s[0].isupper():
        return False
    
    return s != s.lower() and s != s.upper() and "_" not in s
```

```python
boun = pd.read_csv('boun_hashformers_output.csv')
boun
```

```python
camel_case = 0
for index, row in boun.iterrows():
    if is_camel_case(row['hashtag']):
        camel_case += 1
camel_case
```
